### Packages and Parameters

```python
from rocrate.rocrate import ROCrate
from rocrate.model.person import Person
from rocrate.model.contextentity import ContextEntity
```

```python
path_s="./sources"
path_o="."
orcid_id = "horcid:0000-0000-0000-0000"
orcid_id2 = "horcid:0000-0000-0000-1111"
institution_name="Area"
```

### Initialize the crate

```python
crate = ROCrate()
crate.name = "Ex1_RO_Crate"
crate.description = "RO_Crate first exercise"
```

### Payload the files

```python
data_file = crate.add_file(path_s+"/"+'nffa_instruments.csv')
script_file = crate.add_file(path_s+"/"+'load_script.py')
paper = crate.add_dataset(path_s+"/"+"CNR_ISM")
```

### Create objects

```python
person = crate.add(Person(crate, orcid_id, properties={
    "@type": "Person",
    "name": "Brazzo",
    "affiliation": "Area"
}))
person2 = crate.add(Person(crate, orcid_id2, properties={
    "@type": "Person",
    "name": "Erica",
    "affiliation": "Area"
}))


affiliation= crate.add(ContextEntity(crate, "Area", properties={
    "@type": "institute",
    "name": "Area",
    "location": "Trieste"
}))
```

### Assign  relations

```python
paper.append_to("author", person)
data_file.append_to("author",person)
script_file.append_to("author",person2)
person.append_to("affiliation",affiliation)
person2.append_to("affiliation",affiliation)
```

### Assign main entry

```python
crate.mainEntity = script_file # Staring point
```

### Save

```python
crate.write("exercise_ro_crate")
crate.write_zip('exercise_ro_crate.zip') 
print("RO-Crate creato!")
```
