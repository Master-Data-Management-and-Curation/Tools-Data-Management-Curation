### Packages

```python
from rdflib import Graph, URIRef, Literal, Namespace
from rdflib.namespace import RDF, RDFS, XSD
```

### Initialize

```python
g = Graph()

EX = Namespace("http://example.org/")
g.bind("ex", EX)
```

### Defining entities

```python
person = URIRef("http://example.org/person/rossi")
org    = URIRef("http://example.org/org/are")
```

### Defining properties and relations

```python
g.add((person, RDF.type, EX.Person))
g.add((person, EX.affiliation, org))
g.add((org, RDF.type, EX.Organization))
g.add((org, RDFS.label, Literal("Area", datatype=XSD.string)))
```

### Save the RDF in the XML format

```python
g.serialize(destination="rdf_example.rdf", format="xml")
```
